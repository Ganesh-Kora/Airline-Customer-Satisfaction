# 7) Create a function to display precision score, recall score, accuracy, classification report, confusion matrix, F1 Score.

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix, classification_report
```

```python
def show_data(model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train)

    # First Print Accuracy Score Like....
    print("\n", "The Accuracy Score Is : ", accuracy_score(y_test, model.predict(x_test)))

    # Then Print The Recall Score...
    print("\n", "The Recall Score Is : ", recall_score(y_test, model.predict(x_test)))

    # Then Print The Precision Score...
    print("\n", "The Precision Score Is : ", precision_score(y_test, model.predict(x_test)))

     # Also We Can See The Confusion Matrix....
    print("\n", "The Confusion Matrix Looks Like Following....")
    print(confusion_matrix(y_test, model.predict(x_test)))

    # And Finally If The Matrix Looks Good Then Its Classification Report Looks Like Following...
    print("\n", "The Classification Report Is Like Following....")
    print(classification_report(y_test, model.predict(x_test)))
```

```python
show_data(model, x_train, y_train, x_test, y_test)
```
